```python
import pandas as pd
# create dataframe
mcdonalds = pd.read_csv("menu.csv")
# display a preview
mcdonalds.head()
```

```python
# print the number of cols, rows
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\nThere are " + str(mcdonalds.shape[1]) + " columns and " + str(mcdonalds.shape[0]) + " rows.\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
```

```python
# basic statistics for the numerical columns
mcdonalds.describe()
```

```python
# show the data types
mcdonalds.dtypes
```

```python
# print all the column names
mcdonalds.columns
```

```python
# display the number of unique values in each categorical column

# set up a list of only the `object` dtypes
object_columns = mcdonalds.select_dtypes(include=['object']).columns

# iterate thru list and print unique() for each
for column in object_columns:
    unique_count = mcdonalds[column].nunique()
    print(f"There are {unique_count} different values in `{column}`.")
```

```python
# check for missing values
missing_values = mcdonalds.isnull().sum()

# calculate missing value counts
missing_values_df = pd.DataFrame(missing_values[missing_values > 0], columns=['Count Missing'])

# calculate missing percent of values for each column
missing_pct = ((missing_values/mcdonalds.shape[0]*100).round(3))

# add missing count and percent to table
missing_values_df["% Missing"] = missing_pct

# if there are no missing values, print this.
if missing_values_df.empty == 0:
    print("There are no missing values. ")
    
# if there are missing values, display the table
else:
    missing_values_df
```

```python
# check for duplicate rows
duplicate_rows = mcdonalds.duplicated().sum()
print("There are " + str(duplicate_rows) + " duplicate row(s).\n")

# get a boolean series indicating which rows are duplicates (including the original rows)
duplicate_mask = mcdonalds.duplicated(keep=False)

# use mask to filter and display both the original and duplicate rows
duplicate_rows_df = mcdonalds[duplicate_mask]
duplicate_rows_df
```
